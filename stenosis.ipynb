```python
import subprocess

# Define your project parameters
input_file = "model.vtp"
coords = "-12.2661 11.9268 -6.9170"
stenosis_levels = [30, 50, 90]

for level in stenosis_levels:
    output_file = f"stenosis_{level}.vtp"
    print(f"--- Generating {level}% Stenosis: {output_file} ---")
    
    # Building the command exactly like the CMD version
    command = [
        "python", "-m", "morphman.manipulate_area",
        "-i", input_file,
        "-o", output_file,
        "-m", "stenosis",
        "-p", str(level),
        "-r", "commandline",
        "-rp", "-12.2661", "11.9268", "-6.9170",
        "--size", "4"
    ]
    
    # Running the command
    result = subprocess.run(command, capture_output=True, text=True)
    
    if result.returncode == 0:
        print(f"Successfully created {output_file}")
    else:
        print(f"Error creating {level}% stenosis:")
        print(result.stderr)

print("\nAll tasks complete. Check your folder for the .vtp files!")
```
