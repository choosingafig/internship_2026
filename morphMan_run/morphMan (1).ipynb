```python
import morphman
import os

# --- 1. Configuration ---
input_mesh = "healthy_coronary.vtp"
centerline_file = "model_centerline.vtp" 
output_folder = "stenosis_models"
center_point = [-12.2661, 11.9268, -6.9170] 

# Degrees for  comparison study
stenosis_degrees = [50, 75, 90]

# Create output directory if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

print("Starting Stenosis Generation...\n")

# --- 2. Processing ---
for degree in stenosis_degrees:
    output_filename = os.path.join(output_folder, f"coronary_stenosis_{degree}.vtp")
    print(f"--- Processing {degree}% Stenosis ---")
    
    try:
        
        morphman.manipulate_area(
    input_mesh,           # 1. input_filepath
    "stenosis",           # 2. method (was hardcoded "centerline" → change to "stenosis")
    True,                 # 3. smooth
    0.5,                  # 4. smooth_factor
    False,                # 5. no_smooth
    [],                   # 6. no_smooth_point (empty list)
    "commandline",        # 7. region_of_interest
    center_point,         # 8. region_points (your center point as list)
    0.0,                  # 9. beta
    0.0,                  # 10. ratio
    4.0,                  # 11. size
    float(degree),        # 12. percentage
    output_filename,      # 13. output_filepath
    0.0,                  # 14. poly_ball_size (scalar → use [0.0, 0.0, 0.0] if needed, but doc uses scalar)
    0.1,                  # 15. resampling_step
    0.0,                  # 16. angle_asymmetric
    1.0                   # 17. transition_length
)
        
        # Verify if file was actually created and is not empty
        if os.path.exists(output_filename) and os.path.getsize(output_filename) > 0:
            print(f"Successfully saved: {output_filename}")
        else:
            print(f"Warning: Command finished but {output_filename} is empty or missing.")

    except Exception as e:
        print(f"Error generating {degree}% stenosis: {e}")

print("\n--- Process Finished ---")
```
